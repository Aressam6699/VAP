```python
#https://www.youtube.com/watch?v=mHl5P-qlnCQ&list=PLBv09BD7ez_6cgkSUAqBXENXEhCkb_2wl
#https://www.youtube.com/watch?v=vuc93jbO2Dw
#http://www.cs.cornell.edu/~tomf/publications/supervised_kmeans-08.pdf (supervised clustering)
#https://stackabuse.com/the-best-machine-learning-libraries-in-python/ 
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import ndimage
from sklearn import cluster
image = ndimage.imread("pavan.jpg")
plt.figure(figsize = (15,8))
plt.imshow(image)
```

```python
import numpy as np
```

```python
image.shape
```

```python
x, y, z = image.shape
image_2d = image.reshape(x*y, z)
image_2d.shape
```

```python
kmeans_cluster = cluster.KMeans(n_clusters=8)
kmeans_cluster.fit(image_2d)
cluster_centers = kmeans_cluster.cluster_centers_
cluster_labels = kmeans_cluster.labels_
```

```python
plt.figure(figsize = (15,8))
plt.imshow(cluster_centers[cluster_labels].reshape(x, y, z).astype(np.uint8))
```
